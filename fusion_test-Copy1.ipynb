```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
filepath = "/home/sai/Desktop/TFT_fusion/data/CF_physics_Jan20,23_a.csv"
data = pd.read_csv(filepath)
```

```python
data
```

```python
data["PROB"] = data["YES"]/(data["YES"]+data["NO"])
```

```python
data["METRIC"] = (data["YES"]+data["NO"]) * (data["YES"]-data["NO"])
```

```python
pivoted_data = data.pivot(index ="YEAR", columns="SYSTEM", values="METRIC")
pivoted_data
```

```python
def quantile_normalize(df):
    """
    input: dataframe with numerical columns
    output: dataframe with quantile normalized values
    """
    df_sorted = pd.DataFrame(np.sort(df.values,
                                     axis=0),
                             index=df.index,
                             columns=df.columns)
    df_mean = df_sorted.mean(axis=1)
    df_mean.index = np.arange(1, len(df_mean) + 1)
    df_qn = df.rank(method="min").stack().astype(int).map(df_mean).unstack()
    return (df_qn)
```

```python
quant_norm_pivot_data = quantile_normalize(pivoted_data)
```

```python
quant_norm_pivot_data["YEAR"] = quant_norm_pivot_data.index
```

```python
pd.melt(quant_norm_pivot_data, id_vars=['YEAR'], var_name=['SYSTEM'])
```

```python
quant_norm_data = pd.melt(quant_norm_pivot_data, id_vars=['YEAR'], var_name=['SYSTEM'])
quant_norm_data.rename({"value": "QUANT_NORM"}, axis=1, inplace=True)
quant_norm_data.columns
```

```python
pd.merge(quant_norm_data, data, how='inner', on=["YEAR","SYSTEM"])
```
